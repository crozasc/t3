```python
#Cargar el conjunto de datos procesado de la tarea 2 como punto de partida.
import requests
url="https://raw.githubusercontent.com/crozasc/t2/main/New.csv"
target_csv_path="forbes_2022_billionaires.csv"

response= requests.get(url)

response.raise_for_status()
with open(target_csv_path,"wb") as f:
    f.write(response.content)
print("CSV descargado correctamente")
```

```python
# Cargar librerias.
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
# Utilizar los métodos pandas.read_xxx() y pandas.to_xxx() para leer y guardar los datos.
df = pd.read_csv("forbes_2022_billionaires.csv")
df.head()
```

```python
#Grafico 1
mpl.style.use('dark_background')
df.groupby(['category']).size().plot(
    kind = "bar",
    figsize=(13, 10),
    alpha=0.9,
    color=['coral', 'darkslateblue', 'mediumseagreen']
    )
plt.title('Millonaires categories')
plt.ylabel('Number of millonaires')
plt.xlabel('Categories')
plt.legend(labels=df.index, loc='upper left') 
plt.savefig('1.png')
plt.show()
```

```python
#Grafico 2
mpl.style.use('tableau-colorblind10')
df.groupby(['selfMade']).size().plot(
    kind='pie',
    figsize=(5,6),
    autopct='%1.1f%%', 
    colors=['lightcoral','purple'],
    pctdistance=1.12, 
    labels=None
    )
plt.ylabel('')
plt.title('Percentage of millionaires selfmade')
plt.axis('equal') # Establece el gráfico para que parezca un círculo.
plt.legend(["False","True"],loc='upper left') 
plt.rcParams.update({'font.size': 11})
plt.savefig('2.png')
plt.show()
```
